# Using BALD for Intertemporal Predictions

Need to implement tensor flow, gpflow and trieste.

```python
import tensorflow as tf
import gpflow
import trieste
```

```python
import matplotlib.pyplot as plt
import numpy as np
import copy
from trieste.acquisition.function import BayesianActiveLearningByDisagreement as BALD
from trieste.acquisition.rule import OBJECTIVE
from trieste.models.gpflow.models import VariationalGaussianProcess as VGP
from trieste.objectives.utils import mk_observer 
from trieste.models.gpflow.builders import build_vgp_classifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
search_space = trieste.space.Box([-1,-1],[1,1])
input_dim = 2

def circle(x):
    return tf.cast(
        (tf.reduce_sum(tf.square(x), axis=1, keepdims=True) - 0.5) > 0,
        tf.float64,
    )
```

```python
points = 200
X = search_space.sample(points)
observer = mk_observer(circle)
sample_data = observer(X)
x_ = sample_data.query_points.numpy()
y_ = sample_data.observations.numpy()
mask = y_[:,0] == 1
```

```python
plt.figure(figsize=(6,6))
plt.plot(x_[mask,0], x_[mask,1], "oC0", mew=0, alpha=0.5)
_ = plt.plot(x_[np.logical_not(mask), 0], x_[np.logical_not(mask),1], "oC1", mew=0, alpha=0.5)
plt.xlim([-1,1])
plt.ylim([-1,1])
```

## Intertemporal Dataset

Intertemporal dataset. We can either load out-of-domain data or kirby data.

```python
import pandas as pd
```

```python
def load_data(name="charles_oneill"):
    # load training data
    train = pd.read_csv(f"../data/{name}_train.csv")
    cols = ["SIR", "LDR", "Delay", "Answer"]
    # load testing data
    test = pd.read_csv(f"../data/{name}_test.csv")
    return train[cols], test[cols]

def load_kirby(name="charles_oneill"):
    # load training data
    train = pd.read_csv(f"../data/{name}_kirby.csv")
    cols = ["SIR", "LDR", "Delay", "Answer"]
    # load testing data
    test = pd.read_csv(f"../data/{name}_kirby.csv")
    return train[cols], test[cols]
```

We split the data into train and test sets.

```python
data = load_kirby("charles_oneill")[0]
sdr, ldr, delay, answer = data['SIR'], data['LDR'], data['Delay'], data['Answer']
X = np.array([sdr, ldr, delay, answer]).T
X_big = X.copy()
X, test = train_test_split(X, test_size=0.25)
```

```python
def plot_3d(X):
    X_data = X[:,:3]
    answer = X[:,3]
    my_colors = {0:'red',1:'green'}
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(projection='3d')
    
    for i in range(0,X_data.shape[0]):
        ax.scatter(X_data[i,0], X_data[i,1], X_data[i,2],color=my_colors[answer[i]], alpha=0.5)
    
    plt.show()
    
plot_3d(X_big)
```

We choose a set of random points from which to start the algorithm. This is required for model initialsiation.

```python
rands = np.random.randint(X.shape[0], size=(4))
initial_data_set = X[rands].astype(np.float64)
print(initial_data_set.shape)
plot_3d(initial_data_set)
initial_data_set = X[rands].astype(np.float64)[:,:3]
```

## BALD with Gaussian Kernel Model 

We define our model, a means of labeling data and a discrete search space of the available data.

```python
def label(X):
    array = np.zeros(X.shape[0])
    for i,x in enumerate(X):
        index = np.linalg.norm(X_big[:,:3]-x, axis=1).argmin()
        array[i] = X_big[index][3]
    return tf.cast(np.expand_dims(array,axis=1), tf.float64)
```

```python
observer = mk_observer(label)
search_space = trieste.space.DiscreteSearchSpace(tf.cast((X[:,:3]), tf.float64))
init_data = observer(initial_data_set)
x_ = init_data.query_points
y_ = init_data.observations
model = VGP(build_vgp_classifier(init_data, search_space, noise_free=True))
model.update(init_data)
model.optimize(init_data)
acq = BALD()
rule = trieste.acquisition.rule.EfficientGlobalOptimization(acq)
```

```python
# Grid data for later use
x_grid = np.linspace(0,np.max(X_big),20)
xx, yy, zz = np.meshgrid(x_grid,x_grid, x_grid)
Xplot = np.vstack((xx.flatten(), yy.flatten(), zz.flatten())).T
```

```python
from sklearn import svm
from skimage import measure
from sklearn.datasets import make_blobs, make_circles

from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
import matplotlib.cm as cm
import matplotlib as mpl
import matplotlib.patches as mpatches
```

## Training

Here we define how we will train the model. In particular, we update the Gaussian kernel with a new position determined by BALD. We also graph the level sets in 3d.

```python
num_steps = 25
bo = trieste.bayesian_optimizer.BayesianOptimizer(observer,search_space)
data = init_data
prev_model = copy.deepcopy(model)

for step in range(num_steps):
    results = bo.optimize(1,data,model,rule)
    updated_dataset = results.try_get_final_datasets()[OBJECTIVE]
    suggested_loc = updated_dataset.query_points.numpy()[-1,:]
    p, _ = prev_model._model.predict_y(Xplot)
    predictions = p.numpy().reshape(x_grid.shape[0], x_grid.shape[0], x_grid.shape[0])
    
    fig = plt.figure(figsize=(4.5,4.5), dpi=100)
    ax = plt.axes(projection ="3d")
    
    X_data = X_big[:,:3]
    answer = X_big[:,3]
    my_colors = {0:'red',1:'green'}

    for i in range(0,X_data.shape[0]):
        ax.scatter(X_data[i,0], X_data[i,1], X_data[i,2],color=my_colors[answer[i]], alpha=1)
    
    ax.plot(suggested_loc[0], suggested_loc[1], suggested_loc[2], "*b", markersize=30)
    
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    zlim = ax.get_zlim()
    
    levels = np.linspace(np.min(predictions), np.max(predictions), 6)
    
    patches = []
    
    for i, level in enumerate(levels[1:5]):
        my_colors = {0:'red',1:'green', 2:'blue', 3:'yellow', 4:'orange'}
        dx = x_grid[1] - x_grid[0]
        dy = x_grid[1] - x_grid[0]
        dz = x_grid[1] - x_grid[0]
        verts, faces, _, _ = measure.marching_cubes(predictions,level)
        verts *= np.array([dx, dy, dz])
        verts -= np.array([xlim[0], ylim[0], zlim[0]])
        mesh = Poly3DCollection(verts[faces])
        mesh.set_facecolor(cm.RdYlGn(level))
        mesh.set_edgecolor('none')
        mesh.set_alpha(0.3)
        ax.add_collection3d(mesh)
        patches.append(mpatches.Patch(color=cm.RdYlGn(level), label=round(level,2)))

    plt.legend(handles=patches)
    plt.savefig(f'../images/{step}-kirby.pdf', dpi=300, transparent=True, bbox_inches='tight', pad_inches=0)
    plt.show()
    
    x_ = data.query_points
    y_ = data.observations
    ps, _ = prev_model._model.predict_y(test[:,:3])
    a = ps.numpy()
    b = np.expand_dims(test[:,3],axis=1)
    print(1-np.sum(np.abs(np.round(a)-b))/ps.shape[0])
    
    model = results.try_get_final_models()[OBJECTIVE]
    data = updated_dataset
    prev_model = copy.deepcopy(model)
```
